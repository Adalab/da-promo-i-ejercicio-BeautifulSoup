```python
# Importar BeautifulSoup de la biblioteca bs4 para analizar documentos HTML
from bs4 import BeautifulSoup

# Importar funciones o clases del módulo 'soporte' ubicado en el paquete 'src'
from src import soporte as sp

# Importar la biblioteca requests para hacer solicitudes HTTP
import requests

# Importar la biblioteca re para trabajar con expresiones regulares
import re 

# Importar pandas para análisis y manipulación de datos
import pandas as pd
```

```python
# simulación de extraer el HTML con requests.get(url)
# el contenido de esta variable es equivalente al response.content

content = sp.content
```

```python
# creamos la "sopa" con el contenido de la variable content 

sopa = BeautifulSoup(content, "html.parser")

# después de la exploración encontramos un elemento común en el que encontramos todos los datos

nombre_ruta = sopa.findAll("div", {"class": "elementor-widget-container"}) 

# limitamos el output mostrado 
nombre_ruta[:2] 
```

```python
# aplicamos .text en cada elemnto del bucle para obtener el texto de cada elemento
lista_nombres = [ruta.text.strip() for ruta in nombre_ruta]

# exploramos el contenido obtenido
lista_nombres
```

```python
# Hacemos lo mismo para buscar la posión final de nuestra lista, sumamos dos para quedarnos con los elemntos de la descripción de la última ruta

lista_nombres.index("Tres Lagunas\nLa Nava, Barco y Caballeros\nSierra de Gredos sector occidental") + 2
```

```python
#Creamos la nueva lista con los elementos necesarios para nuestra extracción

rutas = lista_nombres[14:53]
```

Al explorar la lista final, podemos encontrar ciertos patrones en los que se encuentran los elementos que buscamos. 
Por una parte tenemos los nombres de las rutas y por otra un string con el resto de datos que estamos buscando.

```python
# Vemos que los nombres de las rutas se encuentran cada 4 posiciones

nombres_rutas = rutas[::4]
```

```python
nombres_rutas
```

```python
# Las descripciones se encuentran desde la posición 2 y de nuevo con un patrón cada 4 posiciones

descripciones = rutas[2::4]
```

```python
# Exploramos el primer elemnto de la lista de descripciones para poer realizar la limpieza del primer elemento y lugo poder hacerlo en bucle

descripciones[0]
```

```python
# Si hacemos split, obtenemos una lista de la que podemos empezar a extraer los datos necesarios

descripciones[0].split("\n")
```

```python
# Creamos un diccionario para almacenar cada uno de los datos de las rutas, los nombres de las rutas ya los tenemos almacenados por orden en una variable

ruta_info = {
    "nombre": nombres_rutas,
    "donde": [],
    "tipo": [],
    "duracion": [],
    "km": [],
    "dificultad": [],
    "esfuerzo": [],
    "descripcion": []
}


# Iteramos por la lista de descripciones 

for descripcion in descripciones:

    # Hacemos split a la descripción para obtener una lista con los elementos por separado y poder aplicar las condiciones
    lista_split = descripcion.split("\n")
    
    # La descripción está siempre en la última posición de la lista siempre
    descripcion = lista_split[-1]
    ruta_info["descripcion"].append(descripcion)

    for detalle in lista_split:

        if "Tipo" in detalle:

            ruta_info["tipo"].append(detalle.split(":")[1])

        elif "Duración" in detalle or "Tiempo" in detalle:

            ruta_info["duracion"].append(detalle.split(":")[1])        
            
        elif "Localización" in detalle or "Donde se realizan" in detalle:

            ruta_info["donde"].append(detalle.split(":")[0])
            
        elif re.match("Km", detalle, re.IGNORECASE):

            ruta_info["km"].append(detalle)
            
        elif re.match("Difi", detalle, re.IGNORECASE):

            ruta_info["dificultad"].append(detalle.split(":")[1].strip())
            
        elif re.match("Esfu", detalle, re.IGNORECASE):
            
            ruta_info["esfuerzo"].append(detalle.split(":")[1].strip())
```

```python
# Finalmente creamos el Dataframe 

df = pd.DataFrame(ruta_info)
df
```
